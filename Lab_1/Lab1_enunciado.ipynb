1- Git es un sistema de control de versiones distribuido, el cual permite trabajar localmente en proyectos, teniendo un respaldo remoto, permitiendo paralelizar el trabajo con otras personas en un mismo proyecto. Por otra parte, un sistema de control de versiones es un tipo de software que permite tener trazabilidad de los cambios realizados en un proyecto, donde a medida que se va editando este, el software toma una instantanea de los archivos, la cual puede ser recuperada mas adelante en caso de ser necesario. Finalmente git se relaciona con el control de versiones ya que este en esta basado en el software descrito anteriormente, solo que git toma un enfoque "peer to peer", donde cada copia local del código es el repositorio completo.

2- 

I) Git es un VCS mientras que github es una página web: github esta diseñado para permitir VCS a través de git, además de funcionar como servidor y foro de ayuda, git es solamente el software VCS con el que opera github.

II) Github mantiene repositorios utilizando almacenamiento en la nube, mientras que git se utiliza localmente en cada máquina: Los repositorios de github se encuentran en un servidor permitiendo que varios desarrolladores puedan trabajar en un mismo proyecto, por otra parte, git es el software que permite trabajar con esos archivos almacenados en la nube

III) Github permite asignar tareas a los distintos colaboradores de un proyecto: Se pueden asignar distintos permisos a los colaboradores del proyecto dependiendo de sus tareas y objetivos, mientras que git solamente se corre localmente en las computadoras y no permite dar permisos a los demas colaboradores.

3- Un hash es una dirección calculada a través de la **hash function** que comunmente compone la estructura 'map de hash'. Esta estructura es común en las bases de datos ya que tiene propiedades deseadas, como el acceso casi constante a una dirección, o el uso reducido de espacio en memoria que utilizan los hash calculados. Al utilizar git, el hash representa una operación de commit. Por ejemplo, al llamar `git log` al momento de esta respuesta, el último commit es 

```

```
Se observa que el commit esta identificado por un hash.

4- Git tiene dos estructuras principales: un index (stage o cache) mutable y un object database inmutable, que solo puede ser modificado por medio de appends.

##### Index: 
El propósito del index es conectar el working directory y el object tree.
##### Object Database: 
>  blob: binary large object, es el contenido de un archivo. Cada blob es la version de un archivo, no tiene metadata asociada.
> tree: el equivalente a un directorio dentro de git. Contiene una lista de de los nombres de los directorios. 
> commit: el commit lleva a cabo el link entre los distintos trees, creando un historial. Sï tiene metadata como por ejemplo un timestamp, log message, etc.
> tag: contiene una referencia a otro objeto y puede tener metadata. Se utiliza principalmente para almancenar una firma digital de un objeto commit asociado a un release de la data siendo trackeada por git.
> packfile: colecciona varios objetos en un archivo comprimido para transporte sobre redes.

5- 
Para crear un repositorio, se utiliza el comando `git init <project name>` en el directorio que se quiere agregar a git. Para agregar elementos al control de versiones, se utiliza `git add <archivo>` o bien `git add .`. Para commitear los cambios al servidor, se utiliza `git commit -m <msg>`. 

Para hacerle seguimiento a un repositorio, se puede utilizar el comando `git status`, que permite saber el status de los archivos locales, si se han agregado nuevos archivos, o si se han modificado archivos y estos no han sido commiteados todavía. El comando `git diff` permite ver los cambios a archivos que no han pasado al stage, hay variaciones de este comando que permiten hacer distintas comparaciones, siendo una de las más útiles, el comparar los cambios entre dos commits. `git log` permite ver el historial de commits: el hash correspondiente a cada commit, junto a su autor y la fecha en que se llevó a cabo el commit. Para obtener los cambios más recientes desde el servidor se utiliza `git fetch`, mientras que si se quiere obtener los cambios más recientes y hacer merge automáticamente se utiliza `git pull`. Finalmente, para llevar los cambios locales al servidor, se utiliza `git push`.

8. Cree una rama nueva y proponga una solución al problema anterior, adjuntando el código corregido y señalando las líneas modificadas. (0.25 pts)

Para crear una rama nueva es necesario abortar el proceso de merge, ```git merge --abort```, esto es solo porque se solicitó crear una rama nueva para hacer la resolución.

Para crear una nueva branch: 

```
git branch resolucion
```
Luego nos movemos al branch resolucion con el comando 
```
git checkout resolucion
```
y la consola señala que hicimos el switch.

En este branch, que es copia del main por defecto, llamamos denuevo al comando `git merge`, donde tendremos el mismo conflicto anterior, pero se harán las modificaciones necesarias para llevar a cabo el merge.
```
git checkout resolucion
Switched to branch 'resolucion'
```
Para solucionar el merge movemos la función 'animales' traída por el merge a main a el archivo `animales.py`. De esta manera, se integra la función `animales` nueva manteniendo la reestructuración que se llevó a cabo en el main (dividir en módulos).

---
`animales.py`
--- 

````python
def animales(input):
    
    '''
    Como hace el animalito?
    '''
    
    if input.lower() == 'gato':
        print('miau')
    elif input.lower() == 'perro':
        print('guau')
    elif input.lower() == 'fox':
        print('https://www.youtube.com/watch?v=jofNR_WkoCE')
    else:
        raise ValueError('animal no reconocido :('))
````

---

````python
from animales import animales

if __name__ == '__main__':
    print("Ingrese el animalito para saber que dice:")
    animal = input()
    animales(animal)
````
