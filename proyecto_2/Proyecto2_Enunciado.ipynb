```python


def objective(trial):
    # Load the Iris dataset for demonstration purposes 
      
    # Define hyperparameters to be tuned
    n_estimators = trial.suggest_int('n_estimators', 80, 150)
    max_depth = trial.suggest_int('max_depth', 6,12 , log=True)
    min_samples_split = trial.suggest_int('n_estimators', 2, 50)
    min_samples_leaf = trial.suggest_int('n_estimators', 2, 50)
    max_features = trial.suggest_categorical('max_features', ['log2'])


    rfc_params = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'max_features': max_features,
    'random_state': 42,
    'class_weight' :dict({0: 1, 1: len(y) / sum(y)})
    }

    rfc_pipeline.named_steps['classifier'].set_params(**rfc_params)
     
    rfc_pipeline.fit(X_train, y_train)

    # Make predictions on the test set and calculate accuracy
    predictions = rfc_pipeline.predict(X_test)
    auc = roc_auc_score(y_test, predictions)

    return auc

 
study = optuna.create_study(direction='maximize')  # maximize accuracy
study.optimize(objective, n_trials=50)  # You can adjust the number of trials

print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial

print("Value: ", trial.value)
print("Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")
```

```python
 
rfc_pipeline.named_steps['classifier'].set_params(**trial.params)

y_pred = rfc_pipeline.predict(X_test)
auc = roc_auc_score(y_test, y_pred)
  
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print("AUC:", auc)
print(classification_report(y_test, y_pred))
```

```python

def objective_lgbm(trial): 
    gbm_params = {
        'max_depth': trial.suggest_int('max_depth', 2, 5),
        'num_leaves': trial.suggest_int('num_leaves', 20, 30),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 1),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 50),
    } 

    # Set LightGBM classifier parameters
    gbm_pipeline.named_steps['classifier'].set_params(**gbm_params)

    # Train the model
    gbm_pipeline.fit(X_train, y_train)

    # Make predictions on the validation set
    predictions_val = gbm_pipeline.predict(X_test)

    # Calculate AUC
    auc_val = roc_auc_score(y_test, predictions_val)

    return auc_val


# Optimize hyperparameters using Optuna
study = optuna.create_study(direction='maximize')  # maximize AUC
study.optimize(objective_lgbm, n_trials=100)  # You can adjust the number of trials

# Print results
print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial
print("Value (AUC): ", trial.value)
print("Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")
```
